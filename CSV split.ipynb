```python
import pandas as pd

newDataFrames = {}

for i, (name, df) in enumerate(otherdataframes.items()):
    
    for seriesName, series in df.items():
        series = series.copy()
        
        if series.dtype == 'object' and series.apply(lambda x: isinstance(x, str)).all():
            series.replace('\\N', None, inplace=True)
            
            unique_count = series.nunique()
            count = series.value_counts()
            total_rows = len(series)

            occurrence_threshold = 5 

            if (count.max() > occurrence_threshold and unique_count < (total_rows / 10)):
                
                unique_values = series.unique()
                new_name = series.name
                new_dataframe = pd.DataFrame(unique_values, columns=[new_name])
                new_dataframe = new_dataframe.reset_index()
                mapping = dict(zip(new_dataframe[new_name], new_dataframe['index']))

                running_df = df.copy()
                running_df[new_name] = running_df[new_name].map(mapping)
                newDataFrames[new_name] = new_dataframe
```
